```python
import sympy as sp
import numpy as np
from matplotlib import pyplot as plt
import mplcyberpunk
plt.style.use("cyberpunk")


x,y,z,alpha,beta=sp.symbols( 'x,y,z,alpha,beta', real=True)
i,j,j=sp.symbols('i,j,k', integer=True, positive=True)
p=sp.plot( x**2, (x,-10,10), legend=True, title='x^2', show=False )
p.extend( sp.plot( x**2-5, (x,-10,10), legend=True,  title='x^2-5', show=False ) )
p.show()
```

```python
def f( x, a, b, c ):
    return a*x**2+b*x+c

xlist=np.linspace( -10,10, num=1000 )
ylist=f( xlist, 3,1, 4)
plt.plot( xlist, ylist )
```
